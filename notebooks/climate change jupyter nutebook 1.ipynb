```python
# Loading Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore") 
```

```python
data = pd.read_csv('../data/Emissions datasets/CO2 emissions (Aggregate dataset (2021)).csv')
```

```python
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 300)
```

```python
data.head()
```

```python
data.Entity.value_counts()
```

```python
world = data[data['Entity'].str.contains('World')]
world.head()
```

```python
years = world['Year'].unique().tolist()
print(len(years))
print("Data is available from {} to {}".format(min(years), max(years)))
```

```python
#Last 100 years
world = world[(world.Year >= 1900)]
```

```python
## Annual Emmisions by Year

plt.figure(figsize=(20,10))

plt.plot(world['Year'].values, world['Annual CO2 emissions'].values)

# Label the axes
plt.xlabel('Year')
plt.ylabel('Annual CO2 emissions rate')

#label the figure
plt.title('Annual CO2 emissions')


plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,10))


coal = world['CO2 emissions from coal']
oil = world['CO2 emissions from oil']
gas = world['CO2 emissions from gas']
year = world['Year']


ax.plot(year, coal, color="blue", label="From coal")
ax.plot(year, oil, color="red", label="From oil")
ax.plot(year, gas, color="green", label="From gas")

# Label the axes
plt.xlabel('Year')
plt.ylabel('Annual CO2 emissions rate')

#label the figure
plt.title('Annual CO2 emissions by source')

leg = ax.legend(prop={'size': 12}, frameon=True)
leg.set_title('Legend', prop={'size': 14, 'weight': 'bold'})
leg.get_frame().set_color('#F2F2F2')
leg.get_frame().set_edgecolor('#000000')
```

```python
###Last 100 years 

data = data[(data.Year >= 1900)]
```

### Different Regions

```python
world = data[data['Entity'] == 'World']
china = data[data['Entity'] == 'China']
india= data[data['Entity'] == 'India']
asia= data[data['Entity'] == 'Asia (excl. China & India)']
europe= data[data['Entity'] == 'Europe']
northamerica = data[data['Entity'] == 'North America (excl. USA)']
usa= data[data['Entity'] == 'United States']
africa = data[data['Entity'] == 'Africa']
southamerica= data[data['Entity'] == 'South America']


conts = [africa, usa, asia, europe, southamerica, northamerica, china, india, world]
continents = pd.concat(conts)
print(continents['Entity'].unique())





```

## G20 countries

```python
g20 = data.query("Entity == ['Argentina', 'Australia', 'Brazil', 'Canada', 'Saudi Arabia','China', 'France', 'Germany', 'India', 'United States','Indonesia', 'Italy', 'Japan', 'Mexico', 'Russia','South Africa', 'South Korea', 'Turkey', 'United Kingdom', 'Spain']")
print(g20['Entity'].unique())
```

```python
continents.head()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='Entity',y='Annual CO2 emissions',data=continents[continents["Entity"] != "World"])
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='Entity',y='Annual CO2 emissions',data=g20)
plt.xticks(rotation=90)
plt.show()
```
